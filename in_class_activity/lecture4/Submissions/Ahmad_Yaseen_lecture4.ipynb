```python
import pandas as pd
data_path = '/content/Wavetronix.csv'
sensor_data = pd.read_csv(data_path)
sensor_data.head()
```

```python
import matplotlib.pyplot as plt
sensor_data['Time'] = pd.to_datetime(sensor_data['Time'])
plt.figure(figsize=(14, 7))
plt.plot(sensor_data['Time'], sensor_data['WAVE-NE-Speed'], label='WAVE-NE-Speed', color='blue')
plt.title('Speed vs. Time for Northeast Direction')
plt.xlabel('Time')
plt.ylabel('Speed (units)')
plt.grid(True)
speed_threshold = 20
below_threshold = sensor_data['WAVE-NE-Speed'] < speed_threshold
plt.scatter(sensor_data['Time'][below_threshold], sensor_data['WAVE-NE-Speed'][below_threshold],
            color='red', label='Below Threshold Speed')
num_events = below_threshold.sum()
plt.legend()
plt.show()
num_events
```

```python
day_1_data_ne = sensor_data[sensor_data['day'] == 1]
day_1_data_ne['Smoothed_NE_Speed_Day1'] = day_1_data_ne['WAVE-NE-Speed'].rolling(window=10, min_periods=1).mean()
plt.figure(figsize=(14, 7))
plt.plot(day_1_data_ne['Time'], day_1_data_ne['Smoothed_NE_Speed_Day1'], label='Smoothed NE Speed - Day 1', color='blue')
plt.title('Smoothed Speed vs. Time for Northeast Direction on Day 1')
plt.xlabel('Time')
plt.ylabel('Speed (units)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
day_2_data_ne = sensor_data[sensor_data['day'] == 2]
day_2_data_ne['Smoothed_NE_Speed_Day2'] = day_2_data_ne['WAVE-NE-Speed'].rolling(window=10, min_periods=1).mean()
plt.figure(figsize=(14, 7))
plt.plot(day_2_data_ne['Time'], day_2_data_ne['Smoothed_NE_Speed_Day2'], label='Smoothed NE Speed - Day 2', color='blue')
plt.title('Smoothed Speed vs. Time for Northeast Direction on Day 2')
plt.xlabel('Time')
plt.ylabel('Speed (units)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
day_3_data_ne = sensor_data[sensor_data['day'] == 3]
day_3_data_ne['Smoothed_NE_Speed_Day3'] = day_3_data_ne['WAVE-NE-Speed'].rolling(window=10, min_periods=1).mean()
plt.figure(figsize=(14, 7))
plt.plot(day_3_data_ne['Time'], day_3_data_ne['Smoothed_NE_Speed_Day3'], label='Smoothed NE Speed - Day 3', color='blue')
plt.title('Smoothed Speed vs. Time for Northeast Direction on Day 3')
plt.xlabel('Time')
plt.ylabel('Speed (units)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.scatter(day_2_data_ne['Time'], day_2_data_ne['Smoothed_NE_Speed_Day2'], color='blue', label='Smoothed NE Speed - Day 2', s=10)
plt.title('Smoothed Speed vs. Time for Northeast Direction on Day 2 (Scatter Plot)')
plt.xlabel('Time')
plt.ylabel('Speed (units)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
speed_below_60 = day_2_data_ne['Smoothed_NE_Speed_Day2'] < 60
speed_at_or_above_60 = day_2_data_ne['Smoothed_NE_Speed_Day2'] >= 60
plt.figure(figsize=(14, 7))
plt.scatter(day_2_data_ne['Time'][speed_below_60], day_2_data_ne['Smoothed_NE_Speed_Day2'][speed_below_60],
            color='red', label='Speed < 60', s=10)
plt.scatter(day_2_data_ne['Time'][speed_at_or_above_60], day_2_data_ne['Smoothed_NE_Speed_Day2'][speed_at_or_above_60],
            color='blue', label='Speed >= 60', s=10)
plt.title('Classified Speed vs. Time for Northeast Direction on Day 2')
plt.xlabel('Time')
plt.ylabel('Speed (units)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
missing_values = day_2_data_ne['WAVE-NE-Speed'].isnull().sum()
if missing_values > 0:
    day_2_data_ne['WAVE-NE-Speed'].interpolate(inplace=True)
day_2_data_ne.drop_duplicates(subset=['Time'], inplace=True)
day_2_data_ne = day_2_data_ne[(day_2_data_ne['WAVE-NE-Speed'] >= 0) & (day_2_data_ne['WAVE-NE-Speed'] <= 120)]
preprocessing_results = {
    "Missing Values": missing_values,
    "Duplicates Removed": not day_2_data_ne.duplicated(['Time']).any(),
    "Unreasonable Values Filtered": not ((day_2_data_ne['WAVE-NE-Speed'] < 0) | (day_2_data_ne['WAVE-NE-Speed'] > 120)).any()
}
preprocessing_results
```

```python
plt.figure(figsize=(14, 7))
plt.plot(day_2_data_ne['Time'], day_2_data_ne['WAVE-NE-Speed'], label='Cleaned NE Speed - Day 2', color='blue')
plt.title('Cleaned Speed Data for NE Direction on Day 2')
plt.xlabel('Time')
plt.ylabel('Speed (units)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
fft_values = np.fft.fft(day_2_data_ne['WAVE-NE-Speed'])
frequencies = np.fft.fftfreq(len(fft_values), d=1)  # d is the sampling period (1 minute)
fft_magnitude = np.abs(fft_values)
plt.figure(figsize=(14, 7))
plt.stem(frequencies, fft_magnitude, 'b', markerfmt=" ", basefmt="-b")
plt.title('Frequency Spectrum of Cleaned NE Speed Data on Day 2')
plt.xlabel('Frequency (cycles per minute)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
!pip install PyWavelets
import matplotlib.pyplot as plt
import numpy as np

import pywt # This line imports the PyWavelets module, which provides the 'pywt' functions.

# use 'list' to get a list of all available 1d demo signals
signals = pywt.data.demo_signal('list')

subplots_per_fig = 5
signal_length = 1024
i_fig = 0
n_figures = int(np.ceil(len(signals)/subplots_per_fig))
for i_fig in range(n_figures):
    # Select a subset of functions for the current plot
    func_subset = signals[
        i_fig * subplots_per_fig:(i_fig + 1) * subplots_per_fig]

    # create a figure to hold this subset of the functions
    fig, axes = plt.subplots(subplots_per_fig, 1)
    axes = axes.ravel()
    for n, signal in enumerate(func_subset):
        if signal in ['Gabor', 'sineoneoverx']:
            # user cannot specify a length for these two
            x = pywt.data.demo_signal(signal)
        else:
            x = pywt.data.demo_signal(signal, signal_length)
        ax = axes[n]
        ax.plot(x.real)
        if signal == 'Gabor':
            # The Gabor signal is complex-valued
            ax.plot(x.imag)
            ax.legend(['Gabor (Re)', 'Gabor (Im)'], loc='upper left')
        else:
            ax.legend([signal, ], loc='upper left')
    # omit axes for any unused subplots
    for n in range(n + 1, len(axes)):
        axes[n].set_axis_off()
plt.show()
```

```python
#!/usr/bin/env python

import matplotlib.pyplot as plt
import numpy as np

import pywt

time, sst = pywt.data.nino()
dt = time[1] - time[0]

# Taken from http://nicolasfauchereau.github.io/climatecode/posts/wavelet-analysis-in-python/
wavelet = 'cmor1.5-1.0'
scales = np.arange(1, 128)

[cfs, frequencies] = pywt.cwt(sst, scales, wavelet, dt)
power = (abs(cfs)) ** 2

period = 1. / frequencies
levels = [0.0625, 0.125, 0.25, 0.5, 1, 2, 4, 8]
f, ax = plt.subplots(figsize=(15, 10))
ax.contourf(time, np.log2(period), np.log2(power), np.log2(levels),
            extend='both')

ax.set_title(f'Nino1+2 Wavelet Power Spectrum ({wavelet})')
ax.set_ylabel('Period (years)')
Yticks = 2 ** np.arange(np.ceil(np.log2(period.min())),
                        np.ceil(np.log2(period.max())))
ax.set_yticks(np.log2(Yticks))
ax.set_yticklabels(Yticks)
ax.invert_yaxis()
ylim = ax.get_ylim()
ax.set_ylim(ylim[0], -1)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import pywt

# Load the image from the uploaded file
image_path = '/content/2018-06-30-19-05.jpg'
image = Image.open(image_path)

# Convert the image to grayscale to simplify processing
image_gray = image.convert('L')

# Convert image to numpy array
image_array = np.array(image_gray)

coeffs = pywt.dwt2(image_array, 'haar')
cA, (cH, cV, cD) = coeffs
fig, axs = plt.subplots(2, 2, figsize=(12, 12))
axs[0, 0].imshow(cA, cmap='gray')
axs[0, 0].set_title('Approximation Coefficients', fontsize=14)
axs[0, 0].axis('off')
axs[0, 1].imshow(cH, cmap='gray')
axs[0, 1].set_title('Horizontal Detail Coefficients', fontsize=14)
axs[0, 1].axis('off')
axs[1, 0].imshow(cV, cmap='gray')
axs[1, 0].set_title('Vertical Detail Coefficients', fontsize=14)
axs[1, 0].axis('off')
axs[1, 1].imshow(cD, cmap='gray')
axs[1, 1].set_title('Diagonal Detail Coefficients', fontsize=14)
axs[1, 1].axis('off')
plt.tight_layout()
plt.show()
```
